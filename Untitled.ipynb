```python
import cv2
import matplotlib.pyplot as plt
```

```python
# READ IMAGES

img = cv2.imread("kidsphoto.jpeg")
print(img.shape)

newImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
print(img[0][0])
```

```python
#Show this image

plt.imshow(img)
plt.show()
```

```python
plt.imshow(newImg)
plt.show()
```

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = model.detectMultiScale(newImg, 1.3,1)
```

```python
faces.shape
```

```python
faces
```

```python
for f in faces:
    x,y,w,h = f
    cv2.rectangle(newImg,(x,y), (x+w, y+h),(0,255,0),2)
    
plt.imshow(newImg)
plt.show()
```
